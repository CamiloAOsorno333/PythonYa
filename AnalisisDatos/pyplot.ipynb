## PYPLOT (GRAFICOS)

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x=np.arange(0,10,0.1) ##en este punto se dan mas puntos, en lugar de 0.5 le damos 0.1
y=np.sin(x)
plt.plot(x,y)
```

```python
x=np.arange(0,10,0.1) ##en este punto se dan mas puntos, en lugar de 0.5 le damos 0.1
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y)
plt.plot(x,z)
```

```python
x=np.arange(0,10,0.1) ##en este punto se dan mas puntos, en lugar de 0.5 le damos 0.1
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y)
plt.plot(x,z)
plt.show() ##para que no aparezca la etiqueta y quede mas estilizada 
```

##PERSONALIZACIÓN DE GRAFICOS 

```python
x=np.arange(0,10,0.1) ##en este punto se dan mas puntos, en lugar de 0.5 le damos 0.1
y=np.sin(x)
z=np.cos(x)
plt.grid() ## Regilla
plt.title('Mi primera grafica')
plt.plot(x,y,label='sin(x)')
plt.plot(x,z,label='cos(x)')
plt.xlabel('Eje X')
plt.ylabel('eje Y')
plt.legend() ##para que se muestren las leyendas 
plt.show() ##para que no aparezca la etiqueta y quede mas estilizada 
```

```python
x=np.arange(0,10,0.1) ##en este punto se dan mas puntos, en lugar de 0.5 le damos 0.1
y=np.sin(x)
z=np.cos(x)
plt.grid() ## Regilla
plt.title('Mi primera grafica')
plt.plot(x,y,label='sin(x)')
plt.plot(x,z,label='cos(x)')
plt.xlabel('Eje X')
plt.ylabel('eje Y')
plt.xlim(0,4) ##para delimitar la muestra de datos 
plt.legend() ##para que se muestren las leyendas 
plt.show() ##para que no aparezca la etiqueta y quede mas estilizada 
```

TIPOS DE GRAFICOS

Scatter Plot (puntos)

```python
x=np.random.random([50])
y=np.random.random([50])
plt.scatter(x,y)
plt.show()
```

Bar Plot (Categorizar datos)

```python
opciones=['si','no']
resultados=[40,50]
plt.bar(opciones,resultados) ## Barras Verticales
plt.show()
```

```python
opciones=['si','no']
resultados=[40,50]
plt.barh(opciones,resultados) ## Barras horizontales
plt.show()
```

Histogramas (Representa la distribución de los datos de acuedo a sus frecuencias)

```python
x=np.random.random([50])
plt.hist(x)
plt.show()
```

```python
x=np.random.random([50])
plt.hist(x,20) ## con ese otro parametro se customisa el número de columnas que representa las frecuencias 
plt.show()
```

Box Plot (grafica de bigotes que muestra como estan distribuidos los datos, la linea nos muestra la media)

```python
plt.boxplot(x)
plt.show()
```

Pie plot 

```python
plt.pie(resultados,labels=opciones)
plt.show()
```
